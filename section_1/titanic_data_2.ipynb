```python
#결손데이터 처리하기 -> Null
titanic_dt.isna().head(5)
```

```python
titanic_dt.isna().sum()
```

```python
titanic_dt['Cabin'] = titanic_dt['Cabin'].fillna('C000')
titanic_dt.head(5)
#inplace = True로 하는 방법도 있음
```

```python
titanic_dt['Age'] = titanic_dt['Age'].fillna(titanic_dt['Age'].mean())
titanic_dt['Embarked'] = titanic_dt['Embarked'].fillna('S')
titanic_dt.isna().sum()
```

```python
def square(a):
    return a**2

lambda_square = lambda x : x**2

print(square(2))
print(lambda_square(2))
```

```python
#여러개의 값을 처리할 때
a = [1,2,3]
squares = map(lambda x : x**2, a)
list(squares)
```

```python
titanic_dt['Name_len'] = titanic_dt['Name'].apply(lambda x : len(x))
titanic_dt[['Name', 'Name_len']].head(5)
```

```python
titanic_dt['Child_Adult'] = titanic_dt['Age'].apply(lambda x : 'Child' if x <=15 else 'Adult')
titanic_dt[['Child_Adult', 'Age']].head(8)
```

```python
#외부 함수를 선언해놓고 그 함수를 lambda안에 사용할 수 있음
def old(x):
    if x >= 50:
        return 'old'
    else:
        return 'young'
    
titanic_dt['old_young'] = titanic_dt['Age'].apply(lambda x : old(x))
titanic_dt[['old_young', 'Age']].head(10)
```
